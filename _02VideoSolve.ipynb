```python
import cv2
```

```python
# 对视频进行帧数剥离 并进行存储
video = cv2.VideoCapture("download\【02-理论课】爬虫的流程？从入门到入狱？_哔哩哔哩_bilibili_merged.mp4")

cnt = 0
while True:
    ret,frame = video.read()
    if not ret:
        break
    if cnt % 300 == 0:
        cv2.imwrite(f'download/images/{cnt}.jpg',frame)

    cnt += 1
print(cnt)
video.release()
cv2.destroyAllWindows()
```
