```python
# 处理txt的数据格式
file = 'E:/DL/data/MP3/1_new_text/0.txt'

txt = ''  # 用于存储处理后的文本内容

# 读取文件
with open(file, 'r') as f:
    # 跳过前两行
    for _ in range(2):
        next(f)

    # 处理每一行，删除换行符并添加到 txt 中
    for line in f:
        # 删除行末的换行符并处理每行内容
        line = line.rstrip('\n')  # 删除每行的换行符
        lines = line.split(' ')    # 假设按空格拆分
        print(lines[3])  # 打印拆分后的第4项

        # 将处理后的行添加到 txt 中（没有换行符）
        txt += lines[3]

# 将处理后的内容写回源文件
with open(file, 'w') as f:
    f.write(txt)

print("处理后的文本已写回源文件。")
```

```python
import os

audio_path = r'E:\DL\data\MP3\pre'  # 确保这是正确的路径

# 遍历指定的文件夹
for folder in os.listdir(audio_path):
    folder_path = os.path.join(audio_path, folder)

    # 确保只处理文件夹，跳过文件，并且文件夹名以 "text" 结尾
    if os.path.isdir(folder_path) and folder.endswith("text"):
        print(f"Processing folder: {folder}")

        # 遍历该文件夹中的所有文件
        for text_file in os.listdir(folder_path):
            text_file_path = os.path.join(folder_path, text_file)

            # 只处理文本文件（假设文件扩展名是 .txt）
            if text_file.endswith(".txt"):
                print(f"Processing text file: {text_file_path}")

                # 读取文本文件并处理
                txt = ''  # 用于存储处理后的文本
                with open(text_file_path, 'r') as f:
                    # 跳过前两行
                    for _ in range(2):
                        next(f)

                    # 将每一行数据添加到 txt 中
                    for line in f:
                        line = line.rstrip('\n')  # 删除每行末尾的换行符
                        lines = line.split(' ')
                        # print(lines)
                        # print(lines[3])
                        txt += lines[3]  # 拼接到 txt 中

                print(txt)
                # 将处理后的文本写回原文本文件
                with open(text_file_path, 'w') as f:
                    f.write(txt)
            # break
                # print(f"Finished processing {text_file_path}")

    # break
```
